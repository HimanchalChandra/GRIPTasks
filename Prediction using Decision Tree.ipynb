## Task :To Create decision tree clasifier for 'iris' dataset and visualise it

```
import numpy as np
import pandas as pd                                          #importing necessary libraries
import seaborn as sns
import sklearn.datasets as datasets
```

```
iris= datasets.load_iris()                                    #Loading the dataset
```

```
df=pd.DataFrame(iris.data,columns=iris.feature_names)                       #forming the iris dataframe
print(df.head(5))
```

```
y=iris.target
print(y)
```

```
sns.pairplot(df,kind='scatter')                                                      #visualising the dataset
```

```
from sklearn.tree import DecisionTreeClassifier
dtree= DecisionTreeClassifier()                                        #defining the decision tree algorithm 
dtree.fit(df,y)
print('Decision Tree Classifier Created')
```

```
from sklearn.externals.six import StringIO  
from IPython.display import Image                                           #importing necessary libraries for graph viz
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names, filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
